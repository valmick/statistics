# Uniform Distribution

```python
from scipy.stats import uniform

data_uniform = uniform.rvs(loc=20,scale=10,size=1000,random_state=271828)
```

```python
plt.plot(data_uniform)
```

```python
plt.hist(data_uniform,bins=100)
plt.xlabel("bins from 20 to 30")
plt.ylabel("Frequency - note there are 10 bins and 1000 overall samples")
```

```python
def sns_displot(data,name):
    ax = sns.distplot(data,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
    ax.set(xlabel=name, ylabel='Frequency')
    return ax
```

```python
sns_displot(data_uniform,"Uniform Distribution")
```

```python
from scipy.stats import norm

data_norm=norm.rvs(loc=0,scale=1,size=1000)
sns_displot(data_norm,"Normal Distribution")
```
